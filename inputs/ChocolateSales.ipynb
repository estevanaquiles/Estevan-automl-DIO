```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#leitura de arquivo CSV
df = pd.read_csv('ChocolateSales.csv')
```

```python
df
```

```python
#tratamento da coluna date para timestamp
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#tratamento da coluna 'Amount', retirada da cifra $ e conversão para float
df['Amount'] = df['Amount'].str.replace('$', '').str.replace(',', '.').astype(float)
```

```python
#criação de label encode para a coluna 'Sales Person'
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sales Person'] = le.fit_transform(df['Sales Person'])
#df
```

```python
df.info()
```

```python
#exporta dataset para csv
df.to_csv('ChocolateSales_tratado.csv', index=False)
```

```python
#group by de vendas por Sales Person
sales_person_sales = df.groupby('Sales Person')['Amount'].sum().reset_index()
```
